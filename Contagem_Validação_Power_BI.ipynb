```python
import pandas as pd
```

```python
df = pd.read_excel('/content/data (17).xlsx')
```

```python
df.tail()
```

```python
df.drop(df.tail(2).index,             #retirando as linhas de filtros utilizados pelo PowerBI
        inplace = True)
df.tail()
```

```python
df['TOTAL_OCORRENCIAS'] = df['ID Ocorrência']+df['Regional']
df.head()
```

```python
valor = len(df['TOTAL_OCORRENCIAS'].unique())   #Total ocorrências
print(f'Total Ocorrências:{valor}')
```

```python
df['TOTAL_VEICULOS'] = df['ID Ocorrência']+df['Veículo']
df.head()
```

```python
valor = len(df['TOTAL_VEICULOS'].unique())   #Total veículos
print(f'Total Veículos: {valor}')
```
